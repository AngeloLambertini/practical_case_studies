```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import dataset as dataframe

original_hr_df = pd.read_excel('HR_Database.xlsx')
```

```python
original_hr_df
```

```python
#descriptive stats of the different variables

original_hr_df.describe()
```

```python
#count of missing values per column

print('Number of missing values : ', original_hr_df.isnull().sum())
```

```python
#count of rows with missing values

original_hr_df.isna().any(axis=1).sum()
```

```python
# remove rows with missing values
nonull_hr_df = original_hr_df.dropna()
```

```python
nonull_hr_df
```

```python
#count duplicate rows

nonull_hr_df_duplicates = nonull_hr_df.duplicated()
print(nonull_hr_df_duplicates.value_counts())
```

```python
#remove duplicate rows

hr_df = nonull_hr_df.drop_duplicates()
hr_df
```

```python
# boxplots of key variables by attrition

# Had employeed who left experienced lower monthly rates?

ax = hr_df.boxplot(column='MonthlyRate', by='Attrition', figsize=[18, 6])
ax.set_ylabel('MonthlyRate')
plt.suptitle("") # Suppress the titles
plt.title("Monthly Rate by Attrition")

# Does it appear to be a difference in age between employees who left and stayed?

ax = hr_df.boxplot(column='Age', by='Attrition', figsize=[18, 6])
ax.set_ylabel('Age')
plt.suptitle("") # Suppress the titles
plt.title("Age by Attrition")

# Does it appear to be a difference in the distance traveled to commute between employees who left and stayed?

ax = hr_df.boxplot(column='DistanceFromHome', by='Attrition', figsize=[18, 6])
ax.set_ylabel('DistanceFromHome')
plt.suptitle("") # Suppress the titles
plt.title("Distance From Home by Attrition")

plt.show()
```

```python
## Had employees who left expressed their dissatisfaction in the survey?

dataForPlot = hr_df.groupby('Attrition').mean()['JobSatisfaction']

ax = dataForPlot.plot(kind='bar', figsize=[10, 6])
ax.set_ylabel('avg.JobSatisfaction')
ax.set_title('Job Satisfaction vs Attrition')
plt.show()

# Do employees levae because it's too hard to get promoted?

dataForPlot2 = hr_df.groupby('Attrition').mean()['YearsSinceLastPromotion']

ax = dataForPlot2.plot(kind='bar', figsize=[10, 6])
ax.set_ylabel('avg.YearsSinceLastPromotion')
ax.set_title('Career Opportunities vs Attrition')
plt.show()
```
